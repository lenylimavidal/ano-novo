```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from IPython.display import display, HTML
from collections import Counter
import seaborn as sns
```

```python
# Crear un DataFrame
data = {
    "Mensagem": [
        "Feliz Ano Novo",
        "Saúde",
        "Sucesso",
        "Prosperidade",
        "Amor",
        "Alegria",
        "Esperança",
        "Paz",
        "Felicidade",
        "Harmonia",
        "Bênçãos",
        "Riquezas",
        "União",
        "Gratidão",
        "Sabedoria",
        "Fortaleza",
        "Renovação",
    ]
}
```

```python
df = pd.DataFrame(data)
```

```python
# Función para aplicar estilos al DataFrame
def style_dataframe(df):
    return df.style.set_properties(
        **{
            "color": "white",
            "background-color": "#007bff",
            "font-weight": "bold",
            "font-size": "18px",
            "border": "2px solid #000",
            "text-align": "center",
        }
    )
```

```python
# Mostrar el DataFrame estilizado
styled_df = style_dataframe(df)
display(HTML("<h3 style='color: #2d572c;'>Desejos de Ano Novo: Um ano abençoado e próspero!</h3>"))
display(styled_df)
```

```python
# Generar texto para la nube de palabras
text = " ".join(df["Mensagem"])
```

```python
# Crear una nube de palabras
wordcloud = WordCloud(
    width=800,
    height=400,
    background_color="white",
    colormap="viridis",
    max_font_size=100
).generate(text)
```

```python
# Mostrar la nube de palabras
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("Feliz Ano Novo 2025", fontsize=16)
plt.show()
```

```python
# Calcular frecuencias ajustadas manualmente para balancearlas
data_ajustada = {
    "Emoção": ["Alegria", "Esperança", "Medo", "Amor", "Ansiedade", "Felicidade", "Sucesso"],
    "Frecuencia": [6, 6, 5, 5, 4, 4, 4]  # Distribución más equilibrada
}
```

```python
df_ajustada = pd.DataFrame(data_ajustada)
```

```python
# Calcular probabilidades ajustadas para 2024
total_palabras_2024 = df_ajustada["Frecuencia"].sum()
df_ajustada["Probabilidade_2024"] = df_ajustada["Frecuencia"] / total_palabras_2024
```

```python
# Predicción para 2025 con un incremento del 5%
total_palabras_2025 = total_palabras_2024 * 1.05
df_ajustada["Predição_2025"] = df_ajustada["Probabilidade_2024"] * total_palabras_2025
```

```python
# Mostrar el DataFrame ajustado
print("Predicción simbólica de emociones para 2025:")
print(df_ajustada)
```

```python
# Crear gráfico de barras para la predicción de emociones en 2025
plt.figure(figsize=(10, 6))
colores = {
    "Alegria": "gold",
    "Esperança": "green",
    "Medo": "red",
    "Amor": "purple",
    "Ansiedade": "blue",
    "Felicidade": "pink",
    "Sucesso": "orange"
}
plt.bar(
    df_ajustada['Emoção'], 
    df_ajustada['Predição_2025'], 
    color=[colores[emo] for emo in df_ajustada['Emoção']]
)

# Agregar etiquetas y título con ajustes
plt.xlabel('Categorias de Emoções')  
plt.ylabel('Frequência em 2024') 
plt.title('Predicción de Frecuencia de Emociones en 2025')

# Mostrar valores encima de las barras
for i, v in enumerate(df_ajustada['Predição_2025']):
    plt.text(i, v + 0.1, f'{v:.2f}', ha='center', va='bottom', fontsize=10)

# Mostrar el gráfico
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
